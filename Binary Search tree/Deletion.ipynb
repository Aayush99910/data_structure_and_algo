```python
from collections import deque
class Node: 
  def __init__(self, value):
    self.value = value 
    self.left = None
    self.right = None
  
class BinarySearchTree: 
  def __init__(self, root: Node): 
    self.root = root

  def insert_node(self, rootNode, value: int):
    # base case 
    if (rootNode == None):
      return Node(value)

    if (rootNode.value < value):
      rootNode.right = self.insert_node(rootNode.right, value)
    else:
      rootNode.left = self.insert_node(rootNode.left, value)

    return rootNode 
  
  def findMinimumNode(self, rootNode) -> Node:
    curr = rootNode 
    while (curr and curr.left): 
      curr = curr.left 
    return curr 

  def remove_node(self, rootNode, value: int):
    # base case 
    if rootNode == None: 
      return None 
    
    # Searching for the node 
    if (value < rootNode.value): 
      rootNode.left = self.remove_node(rootNode.left, value)
    elif (value > rootNode.value):
      rootNode.right = self.remove_node(rootNode.right, value)
    else: 
      if (rootNode.left == None):
        return rootNode.right 
      elif (rootNode.right == None):
        return rootNode.left 
      else: 
        minimumNode = self.findMinimumNode(rootNode.right)
        rootNode.value = minimumNode.value 
        rootNode.right = self.remove_node(rootNode.right, minimumNode.value)
    
    return rootNode
   
  def custom_traverse(self, node):
    if node is not None:
      # Print the parent node first
      print(f"Parent: {node.value}")

      # Print the left child if it exists
      if node.left:
        print(f"Left Child: {node.left.value}")
      else:
        print("Left Child: None")

      # Print the right child if it exists
      if node.right:
        print(f"Right Child: {node.right.value}")
      else:
        print("Right Child: None")

      # Recursively traverse the left subtree
      self.custom_traverse(node.left)
      
      # Recursively traverse the right subtree
      self.custom_traverse(node.right)

  def breathFirstSearch(self, root) -> None:
    queue = deque() 

    if (root):
      queue.append(root)
    
    level = 0 
    while (len(queue) > 0):
      print("Level: ", level)
      for i in range(len(queue)):
        curr = queue.popleft()
        print(curr.value) 

        if (curr.left): 
          queue.append(curr.left)
        if (curr.right):
          queue.append(curr.right)
          
      level += 1
```
