```python
class Node:
    def __init__(self, value):
        self.data = value
        self.next = None

class LinkedList():
    def __init__(self):
        self.head = None
        self.n = 0
    
    # printing the list 
    def __str__(self):
        resultString = ""

        if self.head == None:
            return "Empty List"

        currentNode = self.head 
        while (currentNode.next != None):
            resultString = resultString + str(currentNode.data) + "->"
            currentNode = currentNode.next
        
        resultString += str(currentNode.data)

        return resultString

    # inserting a new node at the beginning of the list
    def insertAtHead(self, value):
        # creating a new node
        newNode = Node(value)

        # creating a connection 
        newNode.next = self.head 

        # reassigning 
        self.head = newNode

        # increamenting the n
        self.n = self.n + 1
    
    # appending a new node at the end of the list
    def append(self, value):
        # checking if the list is empty or not
        if (self.head == None):
            self.insertAtHead(value)
            return 
        
        newNode = Node(value) # creating a new node 
        currentNode = self.head # assigning current node

        # looping until we hit the end node
        while (currentNode.next != None):
            currentNode = currentNode.next # reassigning the current Node

        currentNode.next = newNode # adding the new node at the end
        self.n += 1 # increamenting the n counter
        
    def formatString(self):
        newString = ""

        previousNode = None 
        currentNode = self.head
        nextNode = currentNode.next 

        doubleSymbol = False
        while (nextNode != None):
            # two checks 
            # check if there are two occurences of * or / or */ 
            # check if there is a occurence of * or / 
            if (currentNode.data == "*" or currentNode.data == "/") and (nextNode.data == "*" or nextNode.data == "/"):
                newString += " "
                doubleSymbol = True
            elif(currentNode.data == "*" or currentNode.data == "/") and (previousNode.data == "*" or previousNode.data == "/"):
                newString = newString
            elif(currentNode.data == "*" or currentNode.data == "/"):
                newString += " "
            else:
                if (doubleSymbol):
                    newString += currentNode.data.upper()
                    doubleSymbol = False
                else:
                    newString += currentNode.data
            
            previousNode = currentNode
            currentNode = nextNode
            nextNode = currentNode.next
        
        newString += currentNode.data
        return newString
    

    def createNewFormatedList(self):
        currentNode = self.head # first node

        # looping until the end of the list
        while (currentNode != None):
            # check if the data is "*" or "/" if so then we replace by space
            if (currentNode.data == "*" or currentNode.data == "/"):
                currentNode.data = " "
                
                # check if the consecutive data is "*" or "/" or "*/"
                if (currentNode.next.data == "*" or currentNode.next.data == "/"):
                    currentNode.next.next.data = currentNode.next.next.data.upper() # change the next's next node data to upper
                    currentNode.next = currentNode.next.next # and we skip a node
                
            currentNode = currentNode.next # changing the node 


    
```
