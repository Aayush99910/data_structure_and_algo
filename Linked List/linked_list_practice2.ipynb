Write a python program to remove all duplicate elements from a sorted linked list containing integer data. 

# Example
Input Linked List: 10 10 10 20 20 30 30 40 50

Output Linked List: 10 20 30 40 50

```python
class Node:
    def __init__(self, value):
        self.data = value
        self.next = None

class LinkedList():
    def __init__(self):
        self.head = None
        self.n = 0
    
    # printing the list 
    def __str__(self):
        resultString = ""

        if self.head == None:
            return "Empty List"

        currentNode = self.head 
        while (currentNode.next != None):
            resultString = resultString + str(currentNode.data) + "->"
            currentNode = currentNode.next
        
        resultString += str(currentNode.data)

        return resultString

    # inserting a new node at the beginning of the list
    def insertAtHead(self, value):
        # creating a new node
        newNode = Node(value)

        # creating a connection 
        newNode.next = self.head 

        # reassigning 
        self.head = newNode

        # increamenting the n
        self.n = self.n + 1
    
    # appending a new node at the end of the list
    def append(self, value):
        # checking if the list is empty or not
        if (self.head == None):
            self.insertAtHead(value)
            return 
        
        newNode = Node(value) # creating a new node 
        currentNode = self.head # assigning current node

        # looping until we hit the end node
        while (currentNode.next != None):
            currentNode = currentNode.next # reassigning the current Node

        currentNode.next = newNode # adding the new node at the end
        self.n += 1 # increamenting the n counter

    # Removing duplicates
    def removeDuplicateElement(self):
        currentNode = self.head 
        currentData = currentNode.data 
        dataCurrentNode = currentNode

        while (currentNode != None):
            if (currentNode.data != currentData):
                dataCurrentNode.next = currentNode
                dataCurrentNode = currentNode
                currentData = currentNode.data 
            
            currentNode = currentNode.next

        
  


    
```
