We are creating the object node with data members data and address of next node.

```python
class Node:
    def __init__(self, value):
        self.data = value
        self.next = None
```

We are now creating a object LinkedList where we will use store, insert, create, update new nodes.

```python
class LinkedList:
    def __init__(self):
        self.head = None
        self.n = 0

    # getting the item when the index is given
    def __getitem__(self, index):
        # checking if the index exist or not
        if ( 0 <= index < self.n):
            # logic
            position = 0 
            currentNode = self.head 

            # looping until we get the last node
            while (currentNode != None):
                if (position == index):
                    return currentNode.data
                currentNode = currentNode.next
                position = position + 1
            
                    
        return 'Index out of range'
    
    # calculating the length of the linked list
    def __len__(self):
        return self.n
    
    # deleting item by index
    def __delitem__(self, index):
        previousNode = self.head
        currentNode = previousNode.next
        position = 0

        if (index == 0):
            self.deleteFirstItem()
            return

        if (0 <= index < self.n - 1):
            while (position != index):
                previousNode = currentNode
                currentNode = currentNode.next
                position = position + 1
            
            self.deleteValue(previousNode.data)
        else:
            print("Range error - Index out of range")

    # inserting a new node at the beginning of the list
    def insertAtHead(self, value):
        # creating a new node
        newNode = Node(value)

        # creating a connection 
        newNode.next = self.head 

        # reassigning 
        self.head = newNode

        # increamenting the n
        self.n = self.n + 1

    # printing the list 
    def __str__(self):
        resultString = ""

        if self.head == None:
            return "Empty List"

        currentNode = self.head 
        while (currentNode.next != None):
            resultString = resultString + str(currentNode.data) + "->"
            currentNode = currentNode.next
        
        resultString += str(currentNode.data)

        return resultString
    
    # appending a new node at the end of the list
    def append(self, value):
        # checking if the list is empty or not
        if (self.head == None):
            self.insertAtHead(value)
            return 
        
        newNode = Node(value) # creating a new node 
        currentNode = self.head # assigning current node

        # looping until we hit the end node
        while (currentNode.next != None):
            currentNode = currentNode.next # reassigning the current Node

        currentNode.next = newNode # adding the new node at the end
        self.n += 1 # increamenting the n counter

    # inserting a node in between nodes we take the value of the node where we want to insert this new node
    def insertAfter(self, after, value):
        newNode = Node(value) # creating a new node
        currentNode = self.head # assigning the current node to head
        nextNode = currentNode.next #assigning the next node

        # looping until we find the right node
        while (currentNode.data != after):
            currentNode = nextNode
            # if we reach the last node and we still don't find the item we return 'Item not in the list'
            if (nextNode == None):
                return "Item not in the list"
            nextNode = currentNode.next 
    
        
        # getting the node in the middle
        newNode.next = currentNode.next
        currentNode.next = newNode
        self.n += 1 # increamenting the node by 1

    # clearing the node
    def clear(self):
        self.n = 0
        self.head = None

    # deleting the first item
    def deleteFirstItem(self):
        # checking for any empty list 
        if self.head == None:
            return "No item is in the list"
        
        # reassigning the head to be second item in the list
        currentNode = self.head
        nextNode = currentNode.next
        self.head = nextNode
        self.n = self.n - 1

    # deleting the last item
    def pop(self):
        # if head is None we say list is empty
        if self.head == None:
            return "List is empty"
        
        # assigning the currentNode and nextNode 
        currentNode = self.head
        nextNode = currentNode.next

        # if nextNode is none then that means we only have 1 item in the list and 
        # when you pop that item we are left with and empty list
        if nextNode == None:
            self.deleteFirstItem()
            return 

        # looping until we reach the end of the list
        while (nextNode.next != None):
            currentNode = nextNode
            nextNode = currentNode.next
        
        # assigning the second last node's next to be None
        currentNode.next = None
        self.n = self.n -1

    # deleting by value
    def deleteValue(self, value):
        currentNode = self.head
        nextNode = currentNode.next

        # check if the list is empty
        if (currentNode == None):
            return "Empty list"

        if (currentNode.data == value):
            self.deleteFirstItem()
            return

        # handle deleting
        while(nextNode.data != value):
            # check if the item is not in the list
            if(nextNode.next == None):
                return "Item is not in the list"
            currentNode = nextNode
            nextNode = currentNode.next

        currentNode.next = nextNode.next
        nextNode.next = None 
        self.n = self.n - 1

    # find the index of the given value
    def findIndex(self, value):
        currentNode = self.head
        i = 0

        # check if the list is empty or not
        if (currentNode == None):
            return 'Empty list'

        # check if the item is in the list
        if (currentNode.data == value):
            return i
        
        # looping until we hit the last node
        while (currentNode != None):
            if (currentNode.data == value):
                return i
            currentNode = currentNode.next
            i = i + 1
        
        return 'Item not in the list'
    
    # finds the maximum value in the list and replaces it with the given value
    def findMaxAndReplace(self, value):
        currentNode = self.head

        # checks for empty list
        if (currentNode == None):
            return "Empty list"

        nextNode = currentNode.next
        nodeWithMaxValue = None
        maxValue = 0

        # looping until we hit the last node
        while (nextNode != None):
            if (currentNode.data > maxValue):
                maxValue = currentNode.data
                nodeWithMaxValue = currentNode
            currentNode = nextNode
            nextNode = currentNode.next

        # replacing the node(which has the maximum value) with given value
        nodeWithMaxValue.data = value

    # sum of the odd index 
    def sumOfOddIndex(self):
        currentNode = self.head
        i = 0 # tracking the index
        total = 0 # tracking the total

        # looping until we reach the end
        while (currentNode != None):
            if not(i % 2 == 0):
                total += currentNode.data
            i = i + 1
            currentNode = currentNode.next

        return total    
    
    # reverses the list
    def reverseLinkedList(self):
        currentNode = self.head
        previousNode = None
        nextNode = currentNode.next 

        while(nextNode != None):
            currentNode.next = previousNode
            previousNode = currentNode
            currentNode = nextNode
            nextNode = currentNode.next
        
        currentNode.next = previousNode
        self.head = currentNode


        

                  
```
